```python
#label encoding

from sklearn.preprocessing import LabelEncoder

label=LabelEncoder()

columns_to_encode=['Gender','Education_Level','Preferred_Activities','Location','Favorite_Season']

for column in columns_to_encode :

  df[column]=label.fit_transform(df[column])
```

```python
df.dtypes
```

```python
#assigning input features and output features

x=df.iloc[:,:-1].values
x
```

```python
x.shape
```

```python
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=0)

```

```python
#scaling

from sklearn.preprocessing import MinMaxScaler

scalar=MinMaxScaler()

scalar.fit(x_train)

x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
```

```python
#3 classifier algorithms >> knn ,naive bayes and svm  >> to check whether which is giving us more accuracy

from sklearn.neighbors import KNeighborsClassifier

from sklearn.naive_bayes import BernoulliNB  # we use classifier as BNB because it,s sentimental analysis >> boolean

from sklearn.svm import SVC


#models
knn=KNeighborsClassifier(n_neighbors=7)
bayes=BernoulliNB()
support=SVC()

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report

list=[knn,bayes,support]
```

```python
for i in list :

  print('model name is',i)
  print('*'*100)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('y_pred=',y_pred)
  print('*'*100)
  cm=confusion_matrix(y_test,y_pred)
  print('cm=',cm)
  print('*'*100)
  score=accuracy_score(y_test,y_pred)
  print('accuracy_score=',score)
  print('*'*100)
  report=classification_report(y_test,y_pred)
  print(report)
  print('*'*100)

```
